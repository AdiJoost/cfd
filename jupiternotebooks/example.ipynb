```python
import numpy as np
import matplotlib.pyplot as plt
import lic

from config.rootPath import getRootPath

path = getRootPath()
udataPath = path.joinpath("data", "exampleVectors", "u.dat")
vdataPath = path.joinpath("data", "exampleVectors", "v.dat")

data_u = np.loadtxt(udataPath)
data_v = np.loadtxt(vdataPath)

lic_res = lic.lic(data_v, data_u, length=30, contrast=False)
lic_res = np.flip(lic_res,0)
heat = np.sqrt(data_u**2 + data_v**2)
heat = np.flip(heat,0)
plt.imshow(lic_res, cmap='gray')
plt.imshow(heat, alpha=0.6, cmap=plt.cm.coolwarm)
plt.colorbar()
plt.show()
```
